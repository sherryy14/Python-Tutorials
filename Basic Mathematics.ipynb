# Quadratic equation
Ask user to input the coefficients a ,b ,c of quadratic equation and using the quadratic formula calculate the roots of quadratic equation... 
Consider the following example 


```python
a=int(input('Enter 1st Coeff :'))
b=int(input('Enter 2nd Coeff :'))
c=int(input('Enter 3rd Coeff :'))
```

```python
D=(b**2)-(4*a*c)
```

```python
D
```

```python
x1=(-b+(D)**0.5)/(2*a)
x2=(-b-(D)**0.5)/(2*a)
```

```python
print('The value of first root is: ',x1)
print('The value of 2nd root is : ',x2)
```
